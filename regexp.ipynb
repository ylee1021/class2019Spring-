re package를 불러왔고 그 속에 search 함수를 불러오는 것인데 'abcde'에서 'abc'를 찾는 함수이다. 

regular expression은 사전에 적용이 되지 않는다. string에만 적용이 된다. 
<br>
search함수의 input은 두 개이다!! search는 어디에서 무슨 string을 찾아라. 
<br>


```python
re.search('abc', 'abcde')
```

아래에서 abcde에서 ed를 못 찾으니까 아무것도 안 나올거야

```python
re.search('ed$', 'abcde')
```

ed로 끝나는 단어들만 찾아서 10번째까지만 result로 받자

if 절이 없다고 생각하면 wordlist의 갯수만큼 돌면서 그 숫자만큼 w에 넣어지는 것이다.
if 절을 넣으면 result에 넣어지는 것이 훨씬 적어질 것이다. 
한 단어에서 그 단어의 끝에서 ed가 발견이 되면 맨 앞의 w에 넣어라. 그 중에 첫 10번째를 선택하라. 

wordlist는 현재 list. 'a'가 들어가서 if로 들어가겠지. 'ed'를 각 단어에서 찾아서 안 나오면 안 쌓이겠고 나오면 쌓이겠지. 이 작업이 끝까지 반복. 

첫번째 줄이랑 두번째 줄은 같은 값이겠지만 첫번째 줄에서 [:10]을 제외시키면 모든 항목을 다 포함해서 두번째 줄이랑 다르겠지

```python
result = [w for w in wordlist if re.search('ed$', w)][:10]
print(result[:10])
```

bracket안에 있는 character중 아무거나 충족을 시키는 단어를 받아온다.

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
print(result[:10])
```

소수점을 포함하는 양의 실수

bracket 안에 빼기는 뭐에서 뭐까지. 텍스트가 시작을 하고 0에서 9를 맞추고 +가 있으므로 하나 이상 매치를 시킬 수 있고 .만 넣어주고 싶을때 
backslash
<br> 
[0-9] 하면 숫자 하나이고 +를 붙여야 하나 이상 나와도 되는 것이다. 
<br>
.만 넣어주고 싶을때 backslash를 넣어준다. 그냥 .(점)만 하면 character 하나(한 자리)를 의미하는 거니까. 

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
```

\$ == 실제 달라 특수기호 ; 마지막 달라는 끝을 의미.
<br>
output > C$, US$

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
print(wordlist[:10])
```

{4} : 4번 반복   ex. 1614

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
```

+ ; 하나 이상
- : 그냥 대쉬(-)
[a-z] 소문자
{3,5} : 3개에서 5개 나오는 것.
ex. '10-day', '100-share' (output 보면 하나짜리 숫자는 없다. 하나 이상이라고 했는데도)

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
```

{5,} : 5개 이상 (알파벳이 나와야 되고)
{,6} : 6개 이하로 나오는
ex. 'black-and-white', 'bread-and-butter'

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

# supercalifragilisticexpialidocious 전까지 일단 숙제를 해라.

1번 print : 모음철자를 가져온다. (어떤 모음이 가장 많이 나오는지 알 수 있겠지)
2번 print : (..)은 모음 사이에 있는 character를 찾아라
3번 print : 
4번 print : 

```python
word = 'supercalifragilisticexpialidocious'
result = re.findall(r'[aeiou]', word)
print(result)
result = re.findall('[aeiou](..)[aeiou]', word)
print(result)
result = re.findall('[^aeiou].+[^aeiou]', word)
print(result)
result = re.findall('[^aeiou].+?[^aeiou]', word)
print(result)
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```
