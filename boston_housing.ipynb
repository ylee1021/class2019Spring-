```python
# Tensorflow and tf.keras
import tensorflow 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense

#Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

tensorflow.keras.datasets에서 boston_housing data를 불러온다
<br>
총 4개의 variable에 이 data가 담겨있다.

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')

```

input_shape = [13] : 저번 시간과 비교해서 input이 하나였다면 지금은 13개이므로 숫자를 바꾼다.

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

x_train에는 13개의 column이 들어가고 y_train에는 1개의 column이 들어간다. 
<br>
epochs = 1은 데이타를 한번만 돌리는 것이다. 한번으로 안되니까 1000번 돌린다. output을 보면 loss가 작아지는 것을 볼 수 있다.
<br>
fit은 train(훈련)과 같은 말이다. 
20%는 test를 위해서 쓴다. evaluate은 test.

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
#evaluate은 test할 때 사용한다. fit할 때 쓴 데이타를 쓰지 않는다.
```
